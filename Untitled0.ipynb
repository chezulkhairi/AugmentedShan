```python
import cv2
import numpy as np
%pylab
```

```python
im = cv2.imread('test3.jpg')
imshow(im)
```

```python
kernel = np.array([[1]*3]*3)
dilateerode = cv2.erode(cv2.dilate(im,kernel,iterations=2),kernel,iterations=2)
```

```python
imshow(dilateerode)
```

```python
lapl = cv2.Laplacian(im,32,ksize=3)
```

```python
imshow(lapl*5)
```

```python
blurred = cv2.blur(dilateerode,(100,100))
```

```python
imshow(dilateerode-blurred)
```

```python
gray = cv2.cvtColor(dilateerode,cv2.COLOR_BGR2GRAY)
blurred = cv2.blur(gray,(100,100))
imshow(gray > blurred)
```

```python
thresh = 2
significant = (gray > blurred + thresh)*1 - (gray < blurred - thresh)*1
```

```python
imshow(significant)
```

```python
imshow(cv2.blur(significant,(30,30)))
```

```python
imshow(gray > blurred)
```

```python
imshow(cv2.blur(significant.astype(float),(30,30)))
```

```python
imshow(cv2.Canny(((gray>blurred)*255).astype(uint8),60,60))
```

```python
imshow(significant*128+127)
```

```python
houghdraw = im.copy()
houghdraw = numpy.zeros(im.shape).astype(uint8)
houghedges = cv2.Canny(((gray>blurred)*255).astype(uint8),60,60)

minLineLength = 100
maxLineGap = 10
lines = cv2.HoughLinesP(houghedges,1,np.pi/180,100,minLineLength,maxLineGap)
for x1,y1,x2,y2 in lines[0]:
    cv2.line(houghdraw,(x1,y1),(x2,y2),(0,255,0),2)

imshow(houghdraw)
```
